````python
import cv2
import gradio as gr
import numpy as np

# ì–¼êµ´ ê²€ì¶œê¸° ë¡œë“œ
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
"""
"OpenCVëŠ” BGR í˜•ì‹ì´ë¯€ë¡œ ê·¸ë¼ë””ì˜¤ì—ì„œ ë°›ì•„ì˜¨ RGBë¥¼ BGRë¡œ ë³€í™˜ í•´ì„œ ì²˜ë¦¬í•˜ê¸°"

def detect_and_mosaic(image):
  
    # OpenCVëŠ” BGR í˜•ì‹ì´ë¯€ë¡œ ê·¸ë¼ë””ì˜¤ì—ì„œ ë°›ì•„ì˜¨ RGBë¥¼ BGRë¡œ ë³€í™˜
    image_bgr = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    # ì–¼êµ´ ê°ì§€
    faces = face_cascade.detectMultiScale(image_bgr, scaleFactor=1.1, minNeighbors=4)

    # ê°ì§€ëœ ì–¼êµ´ì— ëª¨ìì´í¬ ì ìš©
    for (x, y, w, h) in faces:

        #ë°”ìš´ë”© ë°•ìŠ¤ ê·¸ë¦¬ê¸°
        cv2.rectangle(image_bgr, (x,y), (x+w, y+h), (255,0,0), 2)

        #ëª¨ìì´í¬ ì ìš©
        face = image_bgr[y:y+h, x:x+w]
        small_face = cv2.resize(face, (16, 16), interpolation=cv2.INTER_LINEAR)  # ì¶•ì†Œ
        mozaic_face = cv2.resize(small_face, (w, h), interpolation=cv2.INTER_NEAREST)  # í™•ëŒ€ (ëª¨ìì´í¬ íš¨ê³¼)
        image_bgr[y:y+h, x:x+w] = mozaic_face  # ì›ë³¸ ì´ë¯¸ì§€ì— ì ìš©

    # BGRì„ ë‹¤ì‹œ ê·¸ë¼ë””ì˜¤ì—ì„œ ì¶œë ¥í•  ìˆ˜ ìˆê²Œ RGBë¡œ ë³€í™˜
    result_image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

    return result_image
"""
def detect_and_mosaic(image_path):
    
    # ì´ë¯¸ì§€ ë¶ˆëŸ¬ì˜¤ê¸° (cv2.imreadëŠ” ê¸°ë³¸ì ìœ¼ë¡œ BGR í˜•ì‹)
    image = cv2.imread(image_path)

    # ì–¼êµ´ ê°ì§€
    faces = face_cascade.detectMultiScale(image, scaleFactor=1.1, minNeighbors=4)

    # ê°ì§€ëœ ì–¼êµ´ì— ëª¨ìì´í¬ ì ìš©
    for (x, y, w, h) in faces:
        # ë°”ìš´ë”© ë°•ìŠ¤ ê·¸ë¦¬ê¸°
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # ëª¨ìì´í¬ ì ìš©
        face = image[y:y+h, x:x+w]
        small_face = cv2.resize(face, (16, 16), interpolation=cv2.INTER_LINEAR)  # ì¶•ì†Œ
        mozaic_face = cv2.resize(small_face, (w, h), interpolation=cv2.INTER_NEAREST)  # í™•ëŒ€ (ëª¨ìì´í¬ íš¨ê³¼)
        image[y:y+h, x:x+w] = mozaic_face  # ì›ë³¸ ì´ë¯¸ì§€ì— ì ìš©

    result_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    return result_image

"""
"OpenCVëŠ” BGR í˜•ì‹ì´ë¯€ë¡œ ê·¸ë¼ë””ì˜¤ì—ì„œ ë°›ì•„ì˜¨ RGBë¥¼ BGRë¡œ ë³€í™˜ í•´ì„œ ì²˜ë¦¬í•œê±° ê·¸ë¼ë””ì˜¤ë¡œ êµ¬í˜„í•  ë•Œ"

Gradio ì¸í„°í˜ì´ìŠ¤ ì„¤ì •
iface = gr.Interface(
    fn=detect_and_mosaic,
    inputs=gr.Image(type="numpy"),  # ì´ë¯¸ì§€ ì…ë ¥ (NumPy ë°°ì—´)
    outputs=gr.Image(type="numpy"),  # ì´ë¯¸ì§€ ì¶œë ¥ (NumPy ë°°ì—´)
    title="ì–¼êµ´ ëª¨ìì´í¬ ì²˜ë¦¬ê¸° ğŸ­",
    description="ì—…ë¡œë“œí•œ ì´ë¯¸ì§€ì—ì„œ ì–¼êµ´ì„ ê°ì§€í•˜ê³  ìë™ìœ¼ë¡œ ëª¨ìì´í¬ë¥¼ ì ìš©í•©ë‹ˆë‹¤."
)
"""

iface = gr.Interface(
    fn=detect_and_mosaic,
    inputs=gr.Image(type="filepath"),  # íŒŒì¼ ê²½ë¡œ ì…ë ¥
    outputs="image",  # ì´ë¯¸ì§€ ì¶œë ¥
    title="ì–¼êµ´ ëª¨ìì´í¬ ì²˜ë¦¬ê¸° ğŸ­",
    description="ì—…ë¡œë“œí•œ ì´ë¯¸ì§€ì—ì„œ ì–¼êµ´ì„ ê°ì§€í•˜ê³  ìë™ìœ¼ë¡œ ëª¨ìì´í¬ë¥¼ ì ìš©í•©ë‹ˆë‹¤."
)

# ì‹¤í–‰
iface.launch()
````

# Azure AI ComputerVision Face + OpenCV Mosaic ê·¸ë¼ë””ì˜¤ êµ¬í˜„í•˜ê¸°

````python
import os
import cv2
import gradio as gr
from azure.ai.vision.face import FaceClient
from azure.ai.vision.face.models import (
    FaceDetectionModel,
    FaceRecognitionModel,
    FaceAttributeTypeDetection03
)
from azure.core.credentials import AzureKeyCredential

# ğŸ” Azure AI Computer Vision ìê²© ì¦ëª… ì…ë ¥ / ai sevice ë¦¬ì†ŒìŠ¤ ì…ë ¥í•´ì•¼í•¨
cog_key = ""
cog_endpoint = ""

# âœ… Azure Face Client ì´ˆê¸°í™”
face_client = FaceClient(endpoint=cog_endpoint, credential=AzureKeyCredential(cog_key))

def detect_faces_and_mosaic(image_path):    # ì–¼êµ´ ê°ì§€ ë° ëª¨ìì´í¬ ì²˜ë¦¬ í•¨ìˆ˜

    print("ğŸ” ì–¼êµ´ ê°ì§€ ì¤‘:", image_path)

    # AI ComputerVision ì–¼êµ´ ê°ì§€ ìš”ì²­
    with open(image_path, "rb") as image_data:
        detected_faces = face_client.detect(
            image_content=image_data.read(),
            detection_model=FaceDetectionModel.DETECTION03,
            recognition_model=FaceRecognitionModel.RECOGNITION04,
            return_face_id=False,
            return_face_attributes=[
                FaceAttributeTypeDetection03.HEAD_POSE,
                FaceAttributeTypeDetection03.BLUR,
                FaceAttributeTypeDetection03.MASK
            ],
        )

    # ì–¼êµ´ ì—†ì„ ê²½ìš° ì›ë³¸ ë°˜í™˜
    if not detected_faces:
        print("ğŸš« ì–¼êµ´ì´ ê°ì§€ë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤.")
        return cv2.imread(image_path)

    # ì´ë¯¸ì§€ ë¡œë“œ
    image = cv2.imread(image_path)

    # ì–¼êµ´ë§ˆë‹¤ ëª¨ìì´í¬ ì²˜ë¦¬ ë° ì •ë³´ ì¶œë ¥
    for i, face in enumerate(detected_faces):
        
        r = face.face_rectangle

        x, y, w, h = r.left, r.top, r.width, r.height

        # ëª¨ìì´í¬ ì²˜ë¦¬
        roi = image[y:y+h, x:x+w]
        small = cv2.resize(roi, (16, 16), interpolation=cv2.INTER_LINEAR)
        mosaic = cv2.resize(small, (w, h), interpolation=cv2.INTER_NEAREST)
        image[y:y+h, x:x+w] = mosaic

        # ì–¼êµ´ ì •ë³´ ì‹œê°í™”
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(image, f"Face {i+1}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # BGR â†’ RGB ë³€í™˜ (Gradio ì¶œë ¥ìš©)
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

def detect_and_mosaic(image_path):

    #haarcascades ì–¼êµ´ì¸ì‹ ëª¨ë¸ë¶ˆëŸ¬ì˜¤ê¸°
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # ì´ë¯¸ì§€ ë¶ˆëŸ¬ì˜¤ê¸° (cv2.imreadëŠ” ê¸°ë³¸ì ìœ¼ë¡œ BGR í˜•ì‹)
    image = cv2.imread(image_path)

    # ì–¼êµ´ ê°ì§€
    faces = face_cascade.detectMultiScale(image, scaleFactor=1.1, minNeighbors=4)

    # ê°ì§€ëœ ì–¼êµ´ì— ëª¨ìì´í¬ ì ìš©
    for (x, y, w, h) in faces:
        # ë°”ìš´ë”© ë°•ìŠ¤ ê·¸ë¦¬ê¸°
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # ëª¨ìì´í¬ ì ìš©
        face = image[y:y+h, x:x+w]
        small_face = cv2.resize(face, (16, 16), interpolation=cv2.INTER_LINEAR)  # ì¶•ì†Œ
        mozaic_face = cv2.resize(small_face, (w, h), interpolation=cv2.INTER_NEAREST)  # í™•ëŒ€ (ëª¨ìì´í¬ íš¨ê³¼)
        image[y:y+h, x:x+w] = mozaic_face  # ì›ë³¸ ì´ë¯¸ì§€ì— ì ìš©

    result_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    return result_image

# Gradio UI êµ¬ì„±
with gr.Blocks() as demo:
    gr.Markdown("# ğŸ–¼ï¸ Azure ì–¼êµ´ ê°ì§€ ë° ëª¨ìì´í¬ ì²˜ë¦¬")
    gr.Markdown("ì´ë¯¸ì§€ë¥¼ ì—…ë¡œë“œí•˜ë©´ Azure Face APIë¥¼ ì‚¬ìš©í•´ ì–¼êµ´ì„ ê°ì§€í•˜ê³  ìë™ìœ¼ë¡œ ëª¨ìì´í¬ë¥¼ ì ìš©í•©ë‹ˆë‹¤.")

    with gr.Tab("Azure AI ComputerVision Face + OpenCV Mosaic"):
        with gr.Row():
            input_image = gr.Image(type="filepath", label="ğŸ“¤ ì´ë¯¸ì§€ ì—…ë¡œë“œ")
            output_image = gr.Image(label="ğŸ“¥ ì²˜ë¦¬ ê²°ê³¼")

        run_button = gr.Button("ğŸ§  ì–¼êµ´ ë¶„ì„ ì‹¤í–‰")
        run_button.click(fn=detect_faces_and_mosaic, inputs=input_image, outputs=output_image)
        input_image.change(fn=detect_faces_and_mosaic, inputs=input_image, outputs=output_image)

    with gr.Tab("only-OpenCV Mosaic"):    
        with gr.Row():
            input_image = gr.Image(type="filepath", label="ğŸ“¤ ì´ë¯¸ì§€ ì—…ë¡œë“œ")
            output_image = gr.Image(label="ğŸ“¥ ì²˜ë¦¬ ê²°ê³¼")

        run_button = gr.Button("ğŸ§  ì–¼êµ´ ë¶„ì„ ì‹¤í–‰")
        run_button.click(fn=detect_and_mosaic, inputs=input_image, outputs=output_image)
        input_image.change(fn=detect_and_mosaic, inputs=input_image, outputs=output_image)

# ì•± ì‹¤í–‰
demo.launch()
````
