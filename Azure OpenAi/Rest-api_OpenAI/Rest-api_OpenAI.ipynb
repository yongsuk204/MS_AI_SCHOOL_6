```python
# ì„¤ì •ê°’ì •ë¦¬
azure_oai_endpoint = os.getenv("AZURE_OAI_ENDPOINT")
azure_oai_api_key = os.getenv("AZURE_OAI_KEY")
azure_oai_deployment = os.getenv("AZURE_OAI_DEPLOYMENT")
azure_search_endpoint = os.getenv("AZURE_SEARCH_ENDPOINT")
azure_search_api_key = os.getenv("AZURE_SEARCH_KEY")
azure_search_index_name = os.getenv("AZURE_SEARCH_INDEX")
azure_search_semanticname = os.getenv("AZURE_SEARCH_SEMANTICNAME")
semantic_configuration = "seoulbike-sementic"
```

```python
import requests

def make_search(prompt):
    end_point = azure_oai_endpoint
    header = {
        'Content-Type': 'application/json',
        'api-key': azure_oai_api_key
    }
    body = {
        "messages": [
            {
                "role": "user",
                "content": prompt
            }
        ],
        "past_messages": 10,
        "temperature": 0.7,
        "top_p": 0.95,
        "max_tokens": 800,
        "azureSearchEndpoint": azure_search_endpoint,
        "azureSearchIndexName": azure_search_index_name,
        "data_sources": [
            {
                "type": "azure_search",
                "parameters": {
                    "endpoint": azure_search_endpoint,
                    "index_name": azure_search_index_name,
                    "semantic_configuration": semantic_configuration,
                    "query_type": "semantic",
                    "role_information": "ë„ˆëŠ” ì„œìš¸ì‹œ ê³µê³µìì „ê±° \"ë”°ë¦‰ì´\" ëŒ€ì—¬ì†Œ ìœ„ì¹˜ë¥¼ ì•Œê³ ìˆëŠ” ì „ë¬¸ê°€ì•¼.",
                    "strictness": 3,
                    "top_n_documents": 5,
                    "authentication": {
                        "type": "api_key",
                        "key": azure_search_api_key
                    }
                }
            }
        ]
    }
    
    response = requests.post(url=end_point, headers=header, json=body)

    if response.status_code == 200:
        response_json = response.json()
        message = response_json["choices"][0]["message"]["content"]
        citations = response_json["choices"][0]["message"]["context"]["citations"]
        
        print("\nğŸ“¢ **ë‹µë³€:**")
        print(message)
        
        print("\nğŸ“Œ **ì¶œì²˜:**")
        for idx, citation in enumerate(citations, start=1):
            print(f"\n*ì¸ìš© {idx}*")
            print(citation["content"])
    else:
        print("âŒ ìš”ì²­ ì‹¤íŒ¨: ìƒíƒœ ì½”ë“œ", response.status_code)
```
